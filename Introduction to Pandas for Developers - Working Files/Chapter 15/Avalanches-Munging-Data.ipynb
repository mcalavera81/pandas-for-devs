```python
import re

def to_inches(orig):
    reg = r"""^(((\d*\.)?\d*)')$|^(((\d*\.)?\d*)")$"""
    txt = str(orig)
    if txt == 'nan':
        return orig
    mo = re.search(reg, txt)
    if mo:
        feet = mo.group(2) or 0
        inches = mo.group(5) or 0
        return float(feet)*12 + float(inches)
    else:
        return None
    
to_inches('18"'), to_inches("2.5'"),  to_inches("2.5'18\"")
```

```python
df['depth_inches'] = df.Depth.apply(to_inches)
```
